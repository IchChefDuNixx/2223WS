```python
import matplotlib.pyplot as plt
import math
import numpy as np
y1 = []
y2 = []
y3 = []
l1 = 0.5
l2 = 1
l3 = l1+l2
a = 5
num=100
for i in np.linspace(0,a,num=num):
    y1.append(l1*math.e**(-l1*i))
    y2.append(l2*math.e**(-l2*i))
    y3.append(l3*math.e**(-l3*i))
plt.plot(np.linspace(0,a,num=num),y1)
plt.plot(np.linspace(0,a,num=num),y2)
plt.plot(np.linspace(0,a,num=num),y3)
plt.show()
```
