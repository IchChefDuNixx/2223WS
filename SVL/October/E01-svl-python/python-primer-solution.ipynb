## Task 1

Implement a function that accepts a number $n$ and returns $\sum_{i=1}^n i$.

```python
def py_func(n):    
    return sum([i+1 for i in range(n)])

def np_func(n): 
    return np.sum(np.arange(n)+1)
```

```python
%%timeit
py_func(999999)
```

```python
%%timeit
np_func(999999)
```

# Task 2

Given two dictionaries, merge (with replacement) the key-value pairs from the second dictionary into the first dictionary.

```python
A = {"key1": 3, "key2": "value"}
B = {"key2": 8, "key3": -3}
```

```python
def func(firstDict, secondDict):
    for i in secondDict:
        firstDict[i] = secondDict[i]
    return firstDict

def func_for_lazy_coders(firstDict, secondDict):
    firstDict.update(secondDict) # modifies 'firstDict' inplace; does not return the result
    return firstDict

print(func(A, B))
print(func_for_lazy_coders(A, B))
```

# Task 3

Copy all elements from the following tuple into a list.

```python
imaTuple = tuple([*"asdlkfjalsdkfjlasdjkf"])
```

```python
def tupleToList(someTuple):        
    return list(someTuple)

def tupleToList2(someTuple):
    myList = []
    for element in someTuple:
        myList += [element]
        # Alternatively: myList.extend(element)
    return myList

imaList = tupleToList2(imaTuple)
print(imaList)
type(imaList)
```

# Task 4

Create a 3x3 identity matrix using numpy. Add all rows of the matrix and return the resulting vector.

```python
def func():
    I = np.identity(3)
    I_row_sum = np.sum(I, axis=0)
    return I_row_sum

func()
```
